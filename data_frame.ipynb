```python
import pandas as pd
from data_frame_info import DataFrameInfo
```

```python
# Load data into df
df = pd.read_csv('loan_payments_data.xlsx')  # Adjust the file path accordingly
print(df)
```

```python
# Create an instance of DataFrameInfo
info_extractor = DataFrameInfo(df)

# Use the methods to extract information
column_types = info_extractor.describe_columns()
print("Column Types:\n", column_types)
```

```python
statistical_values = info_extractor.extract_statistical_values()
print("\nStatistical Values:\n", statistical_values)
```

```python
categorical_columns = ['grade', 'sub_grade', 'home_ownership', 'loan_status', 'purpose']
distinct_values = info_extractor.count_distinct_values(categorical_columns)
print("\nDistinct Values in Categorical Columns:\n", distinct_values)
```

```python
info_extractor.print_dataframe_shape()

null_count = info_extractor.count_null_values()
null_percentage = info_extractor.percentage_null_values()
print("\nNull Values Count:\n", null_count)
print("\nNull Values Percentage:\n", null_percentage)
```
